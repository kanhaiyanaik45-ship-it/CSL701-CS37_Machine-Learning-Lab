```python
#task1:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```python
#task2
df = pd.read_csv("titanic.csv")

print("Dataset Loaded Successfully!")
```

```python
# Task 3:
print("First 10 Rows:")
print(df.head(10))
```

```python
# Task 4:
print("Last 5 Rows:")
print(df.tail(5))
```

```python
# Task 5:
print("Dataset Shape:")
print(df.shape)
```

```python
# Task 6:
print("Dataset Information:")
df.info()
```

```python
# Task 7:
print("Summary Statistics:")
print(df.describe())

# Record mean, minimum and maximum values of Age and Fare
print("\nAge Statistics")
print("Mean:", df["Age"].mean())
print("Minimum:", df["Age"].min())
print("Maximum:", df["Age"].max())

print("\nFare Statistics")
print("Mean:", df["Fare"].mean())
print("Minimum:", df["Fare"].min())
print("Maximum:", df["Fare"].max())
```

```python
# Task 8:
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Task 9:
median_age = df["Age"].median()
df["Age"] = df["Age"].fillna(median_age)

print("Missing Age values filled with median.")
```

```python
# Task 10:
survivors = df[df["Survived"] == 1]

print("Number of Survivors:", len(survivors))
print(survivors.head())
```

```python
# Task 11: Convert Fare column to a NumPy array
fare_array = df["Fare"].to_numpy()

print("Fare NumPy Array:")
print(fare_array)
```

```python
# Task 12: Calculate maximum and average Fare
max_fare = np.max(fare_array)
avg_fare = np.mean(fare_array)

print("Maximum Fare:", max_fare)
print("Average Fare:", avg_fare)
```

```python
#task 13:
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=20, color="skyblue", edgecolor="black")
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#task 14:
plt.figure(figsize=(6,4))
sns.countplot(x="Survived", data=df, palette="Set2")

plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")

plt.show()
```

```python
#task 15:
plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["Fare"], color="red", alpha=0.6)

plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```
